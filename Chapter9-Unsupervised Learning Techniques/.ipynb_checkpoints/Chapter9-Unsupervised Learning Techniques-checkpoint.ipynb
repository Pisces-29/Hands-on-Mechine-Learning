# 聚类
- 聚类：识别相似实例并将其分配给相似实例的集群或组。
- 聚类可用于：客户细分、数据分析、降维技术、异常检测(离群值检测)、半监督学习、搜索引擎、分割图像。

## K-Means

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],
     [-2.8,  1.3]])
blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])
```

```python
X, y = make_blobs(n_samples=2000, centers=blob_centers,
                  cluster_std=blob_std, random_state=7)
# n_features表示每一个样本有多少特征值
# n_samples表示样本的个数
# centers是聚类中心点的个数，可以理解为label的种类数
# random_state是随机种子，可以固定生成的数据
# cluster_std设置每个类别的方差
```

```python
X[:5]
```

```python
plt.figure(figsize=(8,4),dpi=100)
plt.scatter(X[:,0],X[:,1],c=None,s=1)
plt.grid()
plt.show()
```

```python
# 训练一个k-means聚类器
from sklearn.cluster import KMeans
k=5
kmeans=KMeans(n_clusters=5)
```

```python
y_pred=kmeans.fit_predict(X)
```

```python
y_pred[:5]
```

```python
y[:5] # 类别标签不一样，但是分到的类别相同。
```

在聚类的上下文中，实例的标签是该实例被算法分配给该集群的索引：不要与分类中的类标签进行混淆。

Kmeans保留经过训练的实例的标签副本。可以通过labels_实例变量得到该副本。

```python
y_pred is kmeans.labels_
```

查看算法发现的5个集群中心点。

```python
kmeans.cluster_centers_
```

可以很容易的将新实例分配给中心点最接近的集群。

```python
X_new=np.array([[0,2],[3,2],[-3,3],[-3,2.5]])
kmeans.predict(X_new)
```

观察决策边界，绝大多数实例已经正确分配给适当的集群，但少数实例可能贴错了标签。当集群具有不同的直径时，Kmeans算法的性能不是很好，因为将实例分配给集群的时候，它所关心的只是与中心点的距离。

与其将每个实例分配给一个单独的集群（称为硬聚类），不如为每个实例赋予每个集群一个分数（称为软聚类）会更有用。分数可以是实例与中心点之间的距离。相反，它可以是相似性分数，例如高斯径向基函数。KMeans中，transfrom()方法测量每个实例到每个中心点的距离。

```python
kmeans.transform(X_new)
```

X_new的第一个实例与第一个中心点的距离为1.49，与第二个中心点的距离为2.88......如果你有一个高维数据集并以这种方式进行转换，那么最终将得到一个k维数据集。（进行了降维）

### K-Means算法思想
大致思想是从随机放置中心点开始。然后标记实例，更新中心点(取标记后实例的中心值)，标记实例，更新中心点，以此类推，直到中心点停止移动。这个算法能保证在有限数量的步骤内收敛。

尽管算法会收敛，但它可能不会收敛到正确解(即可能会收敛到局部最优值)：这取决于中心点的初始化。

### 中心点初始化方法
- 如果你碰巧知道了中心点应该在哪里（例如你提前运行了另一种聚类算法），则可以将超参数init设置为包含中心点列表的Numpy数组，并将n_init=1.
- 另一种方法是使用不同的随机初始化多次运行算法，并保留最优解。随机初始化的数据量由超参数n_init控制，默认情况下它等于10。这意味着当你调用fit时，前述算法运行10次，sklearn会保留最优解。

最优解使用模型的惯性来判定，即每个实例与其最接近的中心点之间的均方距离。K-Means运行n_init次算法，并保留模型的最低惯性。

```python
# 可以通过inertia_instance变量来访问模型的惯性
kmeans.inertia_
```

```python
# score()方法返回负惯性
kmeans.score(X)
# 因为score方法必须遵循sklearn的越大越好的原则，正数最小即负数最大。
```

sklearn算法默认使用KMeans++的初始化方法。如采用原始方法则设置init=random。

### 寻找最佳聚类数
- 当K越大时，惯性会越来越小，所以惯性不是选择最佳聚类数的指标。
- 当绘制惯性作为集群K的函数曲线时，曲线通常包含一个称为“肘”的拐点。

有一种更精确的方法使用轮廓分数。它是所有实例的平均轮廓系数。实例的轮廓系数等于(b-a)/max(a,b)，其中a是于同一集群中其它实例的平均距离，b是平均最近集群距离（即到下一个最近集群实例的平均距离）。轮廓系数在-1到1之间变化，轮廓系数接近1表示该实例很好的位于其自身的集群中，并且原理其它集群。轮廓系数接近0表示该实例接近一个集群的边界。接近-1表示该实例可能已分配给错误的集群。

```python
from sklearn.metrics import silhouette_score
silhouette_score(X,kmeans.labels_)
```

轮廓图见书。

### K-Means的局限
- 当集群具有不同的大小、不同的密度或者非球形时，KMeans的表现不佳。
- 在这种类型的椭圆簇上，高斯混合模型非常有效。
- 在运行KMeans之间，请先放缩特征，否则集群可能会变形，这样性能会变差。

## 使用聚类进行预处理
聚类是一种非常有效的降维方法，特别是作为有监督学习之前的预处理步骤。

```python
# 类似MNIST的简单数据集
from sklearn.datasets import load_digits
X_digits,y_digits=load_digits(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_digits,y_digits)
```

我们先拟合一个逻辑回归模型。

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(max_iter=10000,n_jobs=-1)
log_reg.fit(X_train,y_train)
```

```python
log_reg.score(X_test,y_test)
```

精度约为95.3%。我们通过KMeans聚类进行降维后是否可以表现得更好。

```python
X_train.shape # 1347个8x8图像
```

```python
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
pipeline=Pipeline([("kmeans",KMeans(n_clusters=50)),("log_reg",LogisticRegression(max_iter=100000))])
pipeline.fit(X_train,y_train)
```

```python
pipeline.score(X_test,y_test)
```

精度约为96.8%，精度有所提高。最好的方法是用网格搜索来选择n_clusters数。

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    "kmeans__n_clusters":[i for i in range(2,100)]
}
pipeline=Pipeline([("kmeans",KMeans()),("log_reg",LogisticRegression(max_iter=100000))])
```

```python
grid_clf=GridSearchCV(pipeline,param_grid,cv=5,verbose=2,n_jobs=-1)
grid_clf.fit(X_train,y_train)
```

```python
grid_clf.best_estimator_
```

```python
grid_clf.best_estimator_.score(X_test,y_test)
```

## 使用聚类进行半监督学习
半监督学习中，我们有很多不带标签的实例，而少有带标签的实例。

```python
# 我们首先在50个实例上训练一个逻辑回归模型。
n_labeled=50
log_reg=LogisticRegression(max_iter=1000)
log_reg.fit(X_train[:n_labeled],y_train[:n_labeled])
```

```python
log_reg.score(X_test,y_test)
```

效果并不是很好。我们将完整数据集聚类成50个集群，每个集群找到最接近中心点的图像。

```python
k=50
kmeans=KMeans(n_clusters=50)
X_digits_dist=kmeans.fit_transform(X_train)
```

```python
X_digits_dist[:2]
```

```python
# 每列是该实例属于某个族群的分数,选择每个族群分数最小的那个实例，即为最靠近中心点的。
representative_digit_ix = np.argmin(X_digits_dist, axis=0)
```

```python
representative_digit_ix
```

```python
X_representative_digits=X_train[representative_digit_ix]
```

```python
plt.figure(figsize=(8, 2))
for index, X_representative_digit in enumerate(X_representative_digits):
    plt.subplot(k // 10, 10, index + 1)
    plt.imshow(X_representative_digit.reshape(8, 8), cmap="binary", interpolation="bilinear")
    plt.axis('off')

plt.show()
```

```python
y_representative_digits = y_train[representative_digit_ix]
# 书上是手动标注，这里仅仅是为了方便。
```

```python
y_representative_digits
```

现在我们有一个只有50个实例的带标签的数据集，但它们不是随机实例，每个实例都是集群的代表图像。来观察性能。

```python
log_reg=LogisticRegression(max_iter=1000)
log_reg.fit(X_representative_digits,y_representative_digits)
log_reg.score(X_test,y_test)
```

精度有所提升。

如果我们将标签传播到同一集群中的其它实例，会怎么样？？这称为标签传播：

```python
y_train_propagated=np.empty(len(X_train),dtype=np.int32)
```

```python
for i in range(k):
    y_train_propagated[kmeans.labels_==i]=y_representative_digits[i]
```

```python
y_train_propagated
```

```python
# 再次训练
log_reg=LogisticRegression(max_iter=100000)
log_reg.fit(X_train,y_train_propagated)
log_reg.score(X_test,y_test)
```

精度更加提高，我们实现了半监督学习。

还有些问题，我们把每个代表性实例的标签传播到同一集群的所有实例中，包括一些集群边界的实例，这些实例更容易被错误标记。如果我们仅传播最接近中心点20%的实例又会如何？

```python
percentile_closest=20
X_cluster_dist=X_digits_dist[np.arange(len(X_train)),kmeans.labels_]
```

```python
X_cluster_dist # 选出类别分数
```

```python
for i in range(k):
    # 找出每个类别的实例
    in_cluster=(kmeans.labels_==i)
    cluster_dist=X_cluster_dist[in_cluster]
    # 在python中计算一个多维数组的任意百分比分位数，此处的百分位是从小到大排列
    # 求第20%分位的数值
    cutoff_distance=np.percentile(cluster_dist,percentile_closest)
    above_cutoff=(X_cluster_dist>cutoff_distance)
    # 将其余分类置为-1
    X_cluster_dist[in_cluster & above_cutoff]=-1
```

```python
partially_propagated=(X_cluster_dist!=-1)
```

```python
partially_propagated
```

```python
X_train_partially_propagated=X_train[partially_propagated]
y_train_partially_propagated=y_train_propagated[partially_propagated]
```

```python
# 再来训练模型
log_reg=LogisticRegression(max_iter=10000)
log_reg.fit(X_train_partially_propagated,y_train_partially_propagated)
```

```python
log_reg.score(X_test,y_test)
```

仅使用50个带标签的实例，我们获得了91%的准确率。

## DBSCAN
- 该算法将集群定义为高密度区域。
- 该算法工作原理：
    - 对于每个实例，该算法都会计算在距它一小段距离ε内有多少个实例。
    - 如果一个实例在其ε邻域内至少包含min_samples个实例(包括自身)，则该实例被视为核心实例。
    - 核心实例附近的所有实例构成一个集群，这个邻域内可能包含其它核心实例。因此一长串核心实力构成一个集群。
    - 任何不是核心实例，且邻居中没有实例的实例都被视为异常。

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=1000,noise=0.05)
```

```python
dbscan=DBSCAN(eps=0.05,min_samples=5)
dbscan.fit(X)
```

labels_实例变量中可以得到所有实例的标签。

```python
dbscan.labels_[:20]
```

某些实例的集群索引等于-1，这意味着该算法将其视为异常。core_sample_indices变量可以得到核心实例的索引。components_可以得到核心实例本身。

```python
len(dbscan.core_sample_indices_)
```

```python
dbscan.components_
```

DBSCAN类具有fit_predict()方法，但它没有predict()方法。之所以做出这个决定是因为不同的分类算法可以更好地完成不同的任务，因此作者决定让用户选择使用哪种算法。

```python
# 让我们训练一个KNN分类器
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_,dbscan.labels_[dbscan.core_sample_indices_])
```

```python
X_new=np.array([[-0.5,0],[0,0.5],[1,-0.1],[2,1]])
```

```python
knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

我们仅在核心实例上训练了分类器。

```python
# 将距离较远的实例设置为异常
y_dist,y_pred_ix=knn.kneighbors(X_new,n_neighbors=1)
y_pred=dbscan.labels_[dbscan.core_sample_indices_][y_pred_ix]
y_pred[y_dist>0.2]=-1
y_pred.ravel()
```
