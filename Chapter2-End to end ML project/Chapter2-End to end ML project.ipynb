# 下载数据

```python
import os
import tarfile
import urllib
```

```python
DOWNLOAD_ROOT="https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH=os.path.join("datasets","housing")
HOUSING_URL=DOWNLOAD_ROOT+"datasets/housing/housing.tgz"
```

```python
def fetch_housing_data(housing_url=HOUSING_URL,housing_path=HOUSING_PATH):
    os.makedirs(housing_path,exist_ok=True)
    tgz_path=os.path.join(housing_path,"housing.tgz")
    urllib.request.urlretrieve(housing_url,tgz_path)
    housing_tgz=tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
import pandas as pd
```

```python
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path=os.path.join(housing_path,"housing.csv")
    return pd.read_csv(csv_path)
```

# 创建测试集
## 随机抽样

```python
# 通常选数据集的20%作为测试集
import numpy as np
def split_train_test(data,test_radio):
    shuffled_indices=np.random.permutation(len(data))# 对序列随机排序
    test_set_size=int(len(data)*test_radio)
    test_indices=shuffled_indices[:test_set_size]
    train_indices=shuffled_indices[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
# 可以通过如下调用来划分训练集和数据集
train_set,test_set=split_train_test(housing,0.2)
```

如下方法如果重新运行又会划分出一个新的训练集和数据集，对复现工作不利

解决方案，每个实例都用一个标识符来决定是否进入测试集

```python
from zlib import crc32
def test_set_check(identifier,test_ratio):
    # crc32循环冗余校验码
    # To generate the same numeric value across all Python versions and platforms use crc32(data) & 0xffffffff.
    # 返回值在[0,2**32-1]之间
    return crc32(np.int64(identifier))& 0xffffffff < test_ratio*2**32
```

```python
def split_train_test_by_id(data,test_radio,id_column):
    ids=data[id_column]
    in_test_set=ids.apply(lambda id_:test_set_check(id_,test_radio))
    print(in_test_set)
    return data.loc[~in_test_set],data.loc[in_test_set]
```

```python
# 给数据集加上索引
housing_with_id=housing.reset_index()
train_set,test_set=split_train_test_by_id(housing_with_id,0.2,"index")
```

## 分层抽样

假设收入中位数是预测房价的一个重要属性。我们希望收入属性上，测试集能够代表整个数据集中各种不同类型的收入。由于收入中位数是一个连续数值，所以我们得创建收入的分类属性。

```python
# 根据直方图的分布，采用pd.cut()创建分类属性
housing["income_cat"]=pd.cut(housing["median_income"],bins=[0.,1.5,3.0,4.5,6.,np.inf],labels=[x for x in range(1,6)])
```

```python
housing["income_cat"].hist()
```

使用sklearn中的StratifiedShuffleSplit进行分层抽样

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
#n_splits：是将训练数据分成train/test对的组数，可根据需要进行设置，默认为10
for train_index,test_index in split.split(housing,housing["income_cat"]):
    print("TRAIN:", train_index, "TEST:", test_index)#获得索引值
    strat_train_set=housing.loc[train_index]
    strat_test_set=housing.loc[test_index]
```

```python
# 删除income_cat属性，将数据集回复原样
for set_ in (strat_train_set,strat_test_set):
    set_.drop("income_cat",axis=1,inplace=True)
```

## 自定义转换器
- 需要为一些诸如自定义清理操作或组合特定属性等任务编写自己的转换器
- 可以通过继承TransformerMixin作为基类得到fit_transform方法
- BaseEstimator见书

```python
from sklearn.base import BaseEstimator,TransformerMixin
```

```python
rooms_ix,bedrooms_ix,population_ix,households_ix=3,4,5,6
```

```python
class CombinedAttributesAdder(BaseEstimator,TransformerMixin):
    def __init__(self,add_bedrooms_per_room=True):
        self.add_bedrooms_per_room=add_bedrooms_per_room
    def fit(self,X,y=None):
        return self
    def transform(self,X,y=None):
        rooms_per_household=X[:,rooms_ix]/X[:,households_ix]
        population_per_household=X[:,population_ix]/X[:,households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room=X[:,bedrooms_ix]/X[:,rooms_ix]
            return np.c_[X,rooms_per_household,population_per_household,bedrooms_per_room]
            # np.c_是按行连接两个矩阵，就是把两矩阵左右相加，要求行数相等。
        else:
            return np.c_[X,rooms_per_household,population_per_household]
```

```python
attr_adder=CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs=attr_adder.transform(housing.values)
```

## 转换流水线
- 许多数据转换的步骤需要按照顺序执行
- pipeline处理最后一个是估算器之外，前面必须是转换器（即有fit_transform方法）,最后一个也可以是转换器
- 调用是按照顺序执行的，前面执行fit_transform，最后一个执行fit

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline = Pipeline([('imputer', SimpleImputer(strategy='median')), (
    'attribs_adder', CombinedAttributesAdder()), ('std_scaler', StandardScaler())],verbose=True)
```

```python
housing_num_tr=num_pipeline.fit_transform(housing_num)
```

```python
list(housing_num)
```

使用ColumnTransformer对所有列进行转换

```python
from sklearn.compose import ColumnTransformer
num_attribs=list(housing_num)
cat_attribs=["ocean_proximity"]
full_pipeline=ColumnTransformer([("num",num_pipeline,num_attribs),("cat",OneHotEncoder(),cat_attribs)])
```

```python
housing_prepared=full_pipeline.fit_transform(housing)
```
